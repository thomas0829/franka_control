```python
from PIL import Image
image_dir = "/home/marius/Projects/polymetis_franka/temp.jpg"
image = Image.open(image_dir).convert('RGB')
image
```

```python
import torchvision.transforms as transforms
transform = transforms.Compose([
    transforms.RandomApply([transforms.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5, hue=0.05)], p=0.5),
    transforms.RandomApply([transforms.RandomCrop(size=(220, 220))], p=0.5),
    transforms.Pad(padding=2),
    transforms.RandomApply([transforms.RandomAffine(degrees=5, shear=5)], p=0.5)
])

transformed_image = transform(image)

transformed_image
```

```python
import matplotlib.pyplot as plt

# Apply the transformations 100 times
transformed_images = [transform(image) for _ in range(100)]

# Create a grid of images
fig, axes = plt.subplots(10, 10, figsize=(20, 20))

for ax, img in zip(axes.flatten(), transformed_images):
    ax.imshow(img)
    ax.axis('off')

plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.show()
```
