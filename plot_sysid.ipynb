```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
real = np.load("real.npy")
sim = np.load("sim.npy")
```

```python
real.shape, sim.shape
```

```python
plt.close()
for i, (c, d) in enumerate(zip(['x', 'y', 'z'], ['red', 'green', 'blue'])):
    plt.plot(real[:,i], label='real '+c, color=d)
    plt.plot(sim[:,0,i], label='sim '+c, color=d, linestyle='dashed')

plt.legend()
```
